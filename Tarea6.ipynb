```python
class NodoArbol: 
  def __init__( self , value , left=None , right=None): #O(4)
    self.data = value #O(1)
    self.left = left #O(1)
    self.right = right #O(1)

class ArbolBB:
    def __init__( self ):#O(1)
      self.__root = None #O(1)

    def insert( self , value ): #O(2)
      if self.__root == None: #O(1)
        self.__root = NodoArbol( value ) #O(1)
      else:
        self.__insert_nodo__( self.__root, value) 
    

    def __insert_nodo__( self, nodo , value ): 
      if nodo.data == value:
          print("Ya existe") #O(1)
          pass
      elif value < nodo.data:
          if nodo.left == None:
              nodo.left = NodoArbol( value )
          else:
              self.__insert_nodo__(nodo.left , value)
      else:
            if nodo.right == None:
              nodo.right = NodoArbol( value )
            else:
              self.__insert_nodo__(nodo.right , value)
    
    # preorden , posorden
    def transversal( self , formato="inorden"):
        if formato == "inorden":
            print("InOrden") #O(1)
            self.recorrido_in(self.__root)
            print()
        elif formato == "preorden":
            print("PreOrden") #O(1)
            self.recorrido_pre(self.__root)
            print()
        else: #post
            print("PosOrden") #O(1)
            self.recorrido_pos(self.__root)
            print()
    
    def recorrido_in( self , nodo):
        if nodo != None:
          self.recorrido_in( nodo.left )
          print( nodo.data , end = "  " ) #O(1)
          self.recorrido_in( nodo.right )
    
    def recorrido_pre( self , nodo):
        if nodo != None:
          print( nodo.data , end = "  " ) #O(1)
          self.recorrido_pre( nodo.left )
          self.recorrido_pre( nodo.right )
    
    def recorrido_pos( self , nodo):
        if nodo != None:
          self.recorrido_pos( nodo.left )
          self.recorrido_pos( nodo.right )
          print( nodo.data , end = "  " ) #O(1)
    
    def buscar( self , value):
        if self.__root == None:
            print("Arbol vacio") #O(1)
            return None
        else:
            return self.busca_nodo( self.__root , value)
    
    def busca_nodo( self , nodo , value ):
        if nodo == None:
            print("No existe") #O(1)
            return None
        elif nodo.data == value:
            print("Encontrado" , nodo.data) #O(1)
            return nodo
        elif value < nodo.data:
            print("Buscar del lado izq") #O(1)
            return self.busca_nodo( nodo.left , value)
        else:
            print("Buscar del lado der") #O(1)
            return self.busca_nodo( nodo.right , value)

    def eliminar(self, value):
         if self.__root is None:
             print("Arbol Vacio")
         else:
             self.__eliminar(value,self.__root,self.__root)

    def __eliminar(self,value,nodo,nodoAn):
        if nodo.data == value:
          if self.isNone(nodo):
            self.delete(nodoAn,value)
          else:
            if nodo.left is None:
              if nodo.right.left is None:
                nodo.data = nodo.right.data
                nodo.right = None
              else:
                tmp = self.__busquedaMayorR__(nodo.right,nodo)
                nodo.data = tmp.left.data
                tmp.left = None
            else:
              if nodo.left.right is None:
                nodo.data = nodo.left.data
                nodo.left = None
              else:
                tmp = self.__busquedaMayorL__(nodo.left,nodo)
                nodo.data = tmp.right.data
                tmp.right = None
          print(f" {value}  (X) ")
        elif value < nodo.data and nodo.left != None:
            self.__eliminar(value,nodo.left,nodo)
        elif value > nodo.data and nodo.right != None:
            self.__eliminar(value,nodo.right,nodo)
        else:
            print("No existe")
        
    def __busquedaMayorL__(self,nodo,an):
        if nodo.right is None:
          return an
        else:
          return self.__busquedaMayorL__(nodo.right,nodo)

    def __busquedaMayorR__(self,nodo,an):
        if nodo.left is None:
          return an
        else:
          return self.__busquedaMayorR__(nodo.left,nodo)

    def isNone(self,nodo):
        if nodo.left == None and nodo.right == None:
          return True
        else:
          return False
      
    def delete(self,nodo,value):
        try:
          if nodo.left.data == value:
            nodo.left = None
          else:
            nodo.right = None
        except:
          nodo.right = None
        
```

```python
abb = ArbolBB()

abb.insert(50)
abb.insert(25)
abb.insert(14)
abb.insert(27)
abb.insert(15)
abb.insert(21)
abb.insert(80)

print("-----------------")#O(1)
abb.transversal("inorden")
abb.transversal("preorden")
abb.transversal("posorden")


print("-----------------")#O(1)
print("Buscar el 25")#O(1)
res = abb.buscar(25)
abb.transversal()
print("-----------------")#O(1)
print("Eliminar el 25")#O(1)
abb.eliminar(25)
print("-----------------")#O(1)
print("Buscar el 25")#O(1)
abb.transversal()
res = abb.buscar(25)
```
