```python
#info = [8,5,9,3]
info = [6568,5,9,3,4,64,342,34,56,72,12,33,3434,54,0,35,4]
# range(<inicio>, <stop>,<incremento>)
# range( 3,0,-1)
# intercambio -->   a , b = b  , a

def burbuja( data ):          #M O(n)    P  O(1)
    tam = len(data)           #M O(n)    P  O(1)
    for tope in range(tam-1,0,-1):   #M O(1)    P O(na)
        for pivote in range(tope):   #M O(1)    P O(na)
            if data[pivote] > data[ pivote +1 ]:
                data[pivote],data[pivote + 1] =data[pivote+1],data[pivote] 
                                     #M O(0)    P O(2n^2)
    return data                      #M O(0)    P O(1)
                                     #M O(n)    P O(n^2)

print (burbuja( info ))
# O(340)  P O(15)
# O(n)    P O(3n)
# O(n)    P O(n^2)
print (len(info))
```

# Insertion Sort
Como cuando juegas cartas


```python
def insertion_sort(data): # M O(n) P O(1)
  print(f"original{data}") # M O(0) P O(1)
  for pivote in range(1,len(data),1): # M O(1) P O(na)
    for index in range(0,pivote+1,1): # M O(1) P O(na)
      if data[pivote] < data[index]: # M O(0) P O(na)
        tmp = data[pivote] # M O(1) P O(n)
        for i in range(pivote,index,-1): # M O(1) P O(na)
          data[i] = data[i-1] # M O(0) P O(n*m)
        data[index] = tmp # M O(0) P O(n)
  return data # M O(0) P O(1)

# M O(n+4) P O(4+2n+n*m)
# M O(n) P O(n^2)

info = [10,51,2,18,4,31,12,5]
print(insertion_sort(info))
```

# **Merge Sort**

```python
#la division o el script
def mergeSort(data): # M O(n) P O(1)
  tam = len(data)    # M O(1) P O(1)
  #si el tamano es igual a 1 estamos en el caso base
  if tam == 1:     # M O(0) P O(na)
    return data    # M O(0) P O(1)

  mid = tam // 2  # M O(1) P O(2)
  #arreglos derecho
  I = mergeSort(data[0:mid:1]) # PARA PENSAR EN CASA! M O(0) P O(log n/2)
  
  #arreglo izquierdo
  D = mergeSort(data[mid :: 1])
  #print(f'{I}\t\t{D}')
  return merge(I,D)

#juntamos izquierda y derecha
def merge(I,D):
  resultado = [0] * (len(I) + len(D))
  i = 0
  d = 0
  r = 0 
  #mientras i estuviera dentro de su arreglo y d tambien 
  while i < len(I) and d < len(D):
    if I[i] < D[d]:
      resultado[r] = I[i]
      i += 1
    else:
      resultado[r] = D[d]
      #aumentamos en 1 los punteros
      d += 1
    r +=1 
    #procesar los faltantes ya sea izquierda o derecha 
  while i < len(I):
      resultado[r] = I[i]
      i+=1
      r+=1
  while d < len(D):
      resultado[r] = D[d]
      d += 1
      r +=1
  return resultado
```
